```python
import pandas as pd
import os

# Settings
RAW_DATA_PATH = r"D:\My Coding\wildfire_intelligence_mlops\data\raw\Wildfire_Dataset.csv"
PROCESSED_DATA_PATH = r"D:\My Coding\wildfire_intelligence_mlops\data\processed\california_wildfire.csv"

# California Bounding Box (Approximate)
LAT_MIN, LAT_MAX = 32.5, 42.0
LON_MIN, LON_MAX = -124.5, -114.0

def process_data():
    print("ðŸ”¥ Starting Data Processing... (This might take a minute)")
    
    # Check if raw file exists
    if not os.path.exists(RAW_DATA_PATH):
        print(f"âŒ Error: File not found at {RAW_DATA_PATH}")
        return

    chunk_size = 100000  # Read 100k rows at a time to save RAM
    chunks = []
    
    # Read in chunks
    for chunk in pd.read_csv(RAW_DATA_PATH, chunksize=chunk_size):
        # Filter for California Coordinates
        # Column names based on Kaggle dataset (FireCastRL)
        if 'navigationlatitude' in chunk.columns:
            cali_chunk = chunk[
                (chunk['navigationlatitude'] >= LAT_MIN) & 
                (chunk['navigationlatitude'] <= LAT_MAX) &
                (chunk['navigationlongitude'] >= LON_MIN) & 
                (chunk['navigationlongitude'] <= LON_MAX)
            ]
            chunks.append(cali_chunk)
    
    # Combine all chunks
    if chunks:
        df_cali = pd.concat(chunks)
        print(f"âœ… Filtered Data Shape: {df_cali.shape}")
        
        # Save to processed folder
        os.makedirs(os.path.dirname(PROCESSED_DATA_PATH), exist_ok=True)
        df_cali.to_csv(PROCESSED_DATA_PATH, index=False)
        print(f"ðŸ’¾ Saved processed data to: {PROCESSED_DATA_PATH}")
        
        # Print columns for sanity check
        print("\nðŸ” Columns available:")
        print(df_cali.columns.tolist())
    else:
        print("âš ï¸ No data found for the specified region.")

```
